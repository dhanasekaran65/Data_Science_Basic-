```python
# JSON Array Example
import json

with open("Student_detail.json", "r") as file:
    student = json.load(file)

for std in student:
    print("ID:", std["ID"])
    print("Name:", std["Name"])
    print("Salary:", std["Salary"])
    print()
```

### list → use index numbers (data[0])
dict → use keys (data["Name"])

```python
# Modify the JSON Data
import json

with open("Student_detail.json", "r") as file:
    data = json.load(file)

# Modify the value
data[0]["Name"] = "Jocker"
data[1]["Salary"] = "50000"

with open("Student_detail.json", "w") as file:
    json.dump(data, file, indent=4)

print("Updated Successfully")
    
```

```python
# Real-Time exaample
import json

students = [
    {"name": "Arun", "marks": 85},
    {"name": "Priya", "marks": 90},
    {"name": "Kumar", "marks": 78}
]
with open("marks.json", "w") as file:
    json.dump(students, file, indent=4)
with open("marks.json", "r") as file:
    data = json.load(file)
for student in data:
    print(student["name"], "-", student["marks"])
```
