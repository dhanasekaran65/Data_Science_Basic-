```python
#Create Database
from pymongo import MongoClient
client =MongoClient("mongodb://localhost:27017")
data_base= client["person_data"]
print("Database Created Successfully")
```

```python
#Create Collection
collection= data_base["person_data"]
```

```python
# Insert one Document
person = {
    "Name": "Dhanasekaran",
    "Roll_No": 721723243009,
    "Department": "AI&DS",
    "College": "P.A College of Engineering and Technology"
}

result = collection.insert_one(person)

print("Data Inserted Successfully")
```

```python
# Insert Multiple Documents
person = [
    {
    "Name":"Subash",
    "Roll_No":"721723243050",
    "Department":"CSC",
    "College":"VIT"
    },
    {
    "Name":"Thangabalaji",
    "Roll_No":"72172324008",
    "Department":"IT",
    "College":"MIT"
    }
]

result = collection.insert_many(person)
print("Multiple Document Inserted Successfully")
```

```python
#View Documents
for data in collection.find():
    print(data)
```

```python
# Find One Document
data = collection.find_one({"Name":"Subash"})
print(data)
```

```python
# Update one Document
collection.update_one({"Name":"Dhanasekaran"},{"$set":{"Department":"ECE"}})
```

```python
# Update Many Documents
collection.update_many(
    {"Name": "Thangabalaji"},
    {"$set": {"Roll_No": "123456"}}
)
collection.update_many(
    {"Name": "Subash"},
    {"$set": {"Department": "AI&DS"}}
)
collection.update_many(
    {"Name": "Dhanasekaran"},
    {"$set": {"College": "KIT"}}
)
print("Documents Updated Successfully")
```

```python
for data in collection.find():
    print(data)
```

```python
# Delete Document
collection.delete_one({"Name":"Subash"})
print("Deleted Successfully")
```

```python
for data in collection.find():
    print(data)
```

```python
# Count Documents
count = collection.count_documents({})
print("Total Documents:",count)
```
